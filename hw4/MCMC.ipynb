```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def target_density(x):
    return stats.norm.pdf(x, 2, np.sqrt(2))

def proposal(x):
    return np.random.normal(x, 1)

# Metropolis-Hastings MCMC sampler
def Sampler(x0, step_num):
    x = x0
    samples = []
    for _ in range(step_num):
        x_prime = proposal(x)
        acceptance_ratio = target_density(x_prime) / target_density(x)

        if np.random.uniform(0, 1) < acceptance_ratio:
            x = x_prime

        samples.append(x)

    return samples

x0 = 0
step_num = 10000
samples = Sampler(x0, step_num)

# Plotting the results
plt.hist(samples,bins=50, density=True, histtype='step', label="MCMC samples")
x_vals = np.linspace(min(samples), max(samples), 1000)
plt.plot(x_vals, target_density(x_vals), label="True Density")
plt.legend()
plt.show()
```
